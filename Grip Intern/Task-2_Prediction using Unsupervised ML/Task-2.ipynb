```python
data = data.drop(['Id'], axis=1)
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:, [0,1,2,3]].values
```

#**Model fitting and predicting**

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
pred = model.fit_predict(x)
```

```python
ypred = model.predict(x)
```

```python
ypred
```

```python
import matplotlib.pyplot as plt
```

Plotting Predicted values

```python
plt.figure(figsize=(15,7))
plt.scatter(x[ypred == 0, 0], x[pred == 0, 1], s = 70, c = 'yellow', label = 'Cluster1')
plt.scatter(x[ypred == 1, 0], x[ypred == 1, 1], s = 70, c = 'blue', label = 'Cluster2')
plt.scatter(x[ypred == 2, 0], x[ypred == 2, 1],s = 70, c = 'green', label = 'Cluster3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids', marker="*")
plt.legend()
```
