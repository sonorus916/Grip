Model =  Linear Regression

Data = http://bit.ly/w-data


#**Data collecction and Pre-Processing**

```python
import pandas as pd
import numpy as np
```

```python
data_url = "http://bit.ly/w-data" #getting data
data = pd.read_csv(data_url)
data.head()
```

```python
data.info() #viewing DataInfo
```

```python
data.shape #data Shape
```

```python
data.isnull().sum() #checking for null values in data
```

#**Data Visualization**

```python
import matplotlib.pyplot as plt
```

```python
x = data.iloc[:, :-1].values  #splitting data in x and y
y = data.iloc[:, 1].values  
```

Plot for given data

```python
plt.plot(x,y,color= "green", linestyle="dashed",marker='o', markerfacecolor="blue", markersize=12)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title("Hours Vs. Scores")
```

#**Model fitting and testing with Visulaizing it.**

```python
from sklearn.model_selection import train_test_split  #splitting data in test and train
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression   #model Import
model=LinearRegression()
```

Fitting model

```python
model.fit(xtrain, ytrain)
```

Model prediction

```python
ypred = model.predict(xtest)
```

getting model predicted line

```python
m = model.coef_
c = model.intercept_

l=m*x + c
```

plot comparing actual, predicted points and line

```python
plt.figure(figsize=(9,6))
plt.scatter(x,y, label="Given data points", s=70)
plt.plot(x, l, color="green", label="Modeled Predicted Line")
plt.scatter(xtest,ypred, label="Predicted data points", marker='*', color='red', s=90)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title("Hours Vs. Scores")
plt.legend()
```

plot between actual and Predicted

```python
plt.figure(figsize=(9,5))
plt.scatter(xtest,ytest, label="Actual data points", s=70)
plt.plot(x, l, color="green", label="Modeled Predicted Line")
plt.scatter(xtest,ypred, label="Predicted data points", marker='*', color='red', s=90)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title("Actual Vs. Predicted Data Points")
plt.legend()
```

```python
data_out=pd.DataFrame({'Actual Score':ytest,'Predicted Score':ypred})
data_out
```

Error Calculations

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(ytest, ypred))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(ytest, ypred))
print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(ytest, ypred)))
```

#**Prediction for 9.25 Hr**
